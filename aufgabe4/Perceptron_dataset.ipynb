```python
def append_ones_collumn(matrix):
    n,m = matrix.shape
    collumn = np.ones((n,1))
    return np.hstack((matrix,collumn))
```

```python
X = np.random.uniform(size=(4,2))
X_new = append_ones_collumn(X)
print(f"{X}\n\n{X_new}")
# funktioniert
```

```python
def get_sorted_data(x_data, y_data):
    m_minus = x_data[y_data == 0]
    m_plus = x_data[y_data == 1]
    return append_ones_collumn(m_plus), append_ones_collumn(m_minus)
```

```python
def o(weights, x):
    product = weights * x
    test_sum = product.sum()
    test_cond = test_sum >= 0
    return int(test_cond)
```

```python
o(np.array([1.0,2.0,3.0]), np.array([0.5,0.5,0.5]))
# funktioniert
```

```python
def get_accuracy(weights, m_plus, m_minus):
    errors = 0
    elements_plus, _ = m_plus.shape
    elements_minus, _ = m_minus.shape
    elements_total = elements_plus + elements_minus
    
    for x in m_plus:
        gamma = o(weights, x) - 1
        errors += abs(gamma)

    for x in m_minus:
        gamma = o(weights, x) - 0
        errors += abs(gamma)
    
    return (elements_total - errors) / elements_total
```

```python

def delta_rule(learning_rate, m_plus, m_minus): 
    
    # Add a maximum number of iterations to avoid an infinite loop
    max_iterations = 1000  

    # create random weights vector
    _, dimensions = m_plus.shape
    start_weights = np.random.uniform(size=dimensions)
    weights = copy.deepcopy(start_weights)

    # Lists to store weight and accuracy changes
    weights_history = [] 
    accuracy_history = [] 

    accuracy_history.append(get_accuracy(weights, m_plus, m_minus))

    for _ in range(max_iterations):

        # bad code \/
        for x in m_plus:
            gamma = o(weights, x) - 1
            delta_weights = -learning_rate * gamma * x
            weights += delta_weights

        for x in m_minus:
            gamma = o(weights, x) - 0
            delta_weights = -learning_rate * gamma * x
            weights += delta_weights
        # better code now \/
        
        # Store a copy of weights for each iteration
        weights_history.append(copy.deepcopy(weights))  
        
        accuracy = get_accuracy(weights, m_plus, m_minus)
        accuracy_history.append(accuracy)

        if (accuracy > 0.999):
            print("Converged successfully.")
            break

    weights_history = np.array(weights_history)

    return learning_rate, start_weights, weights_history, accuracy_history
```

```python
### PLOTTING ###
def f(x,w):
    return -x*(w[0]/w[1])-w[2]/w[1]

def plot(ws, ax1, ax2=None):
    ax1.plot(x_data[y_data == 0, 0], x_data[y_data == 0, 1], 'rs')
    ax1.plot(x_data[y_data == 1, 0], x_data[y_data == 1, 1], 'go')
    for i,w in enumerate(ws):
        ax1.plot([-5,5], [f(-5,w), f(5,w)], label=f"Attempt {i}")
    ax1.set_xlabel("feature 1")
    ax1.set_ylabel("feature 2")
    ax1.set_ylim(-14,4)
    ax1.set_title('Classification Data with 2 classes')
    ax1.legend()

learning_rate = 0.3

m_plus, m_minus = get_sorted_data(x_data, y_data)
learning_rate, start_weights, weights_history, accuracy_history = delta_rule(learning_rate, m_plus, m_minus)
print(start_weights)

fig= plt.figure(figsize=(10,14))
plt.subplot2grid((4, 3), (0, 0), colspan=2).plot(weights_history)
plt.title(f"start weights: {start_weights}, learning_rate: {learning_rate}")
plt.xlabel('Iterations')
plt.ylabel('Weights')
plt.subplot2grid((4, 3), (1, 0), colspan=2).plot(accuracy_history)
plt.xlabel('Iterations')
plt.ylabel('Accuracy')
plot(weights_history, plt.subplot2grid((4, 3), (2, 0), colspan=2, rowspan=2))
fig.subplots_adjust(hspace=0.5)
```
